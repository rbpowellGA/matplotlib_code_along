```python
#import all the usual stuff

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##without an X pyplot will auto generate for you
plt.plot([1,4,6,23])
```

```python
## or you can just put in the x and y

plt.plot([4,3,2,1],[8,6,4,3])
```

```python
##third argument can be used to set line color and shapes of points
plt.plot([1,2,3,4],[1,4,9,16], 'ro')
```

```python
## Use plt.axis to set the x and y ranges for the chart
plt.plot([1,2,3,4],[1,4,9,16], 'ro')
plt.axis((0,6,0,20))
plt.show()
```

```python
##you can plot multiple lines and their shape/color on one line
t = np.arange(0, 5, 0.15)
plt.plot(t,t,'r--', t, t**2, 'bs', t, t**3, 'g^', t, t**4, 'yo')
plt.axis((0,5,0,100))
plt.show()
```
